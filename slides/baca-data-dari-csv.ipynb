```python
import pandas
df = pandas.read_csv('contoh-data-data-mining.csv')
df
```

```python
df = df[['Time', 'Age', 'Sentiment']]
df
```

```python
def getDatePartOnly(dateTime):
    return dateTime.split(' ')[0]
def getTimePartOnly(dateTime):
    return dateTime.split(' ')[1]
    
def getYearOnly(dateTime):
    return getDatePartOnly(dateTime).split('/')[0]
def getMonthOnly(dateTime):
    return getDatePartOnly(dateTime).split('/')[1]
def getDateOnly(dateTime):
    return getDatePartOnly(dateTime).split('/')[2]
    
def getHourOnly(dateTime):
    return getTimePartOnly(dateTime).split(':')[0]
def getMinuteOnly(dateTime):
    return getTimePartOnly(dateTime).split(':')[1]
def getSecondOnly(dateTime):
    return getTimePartOnly(dateTime).split(':')[2]

df['Year'] = df['Time'].apply(getYearOnly)
df['Month'] = df['Time'].apply(getMonthOnly)
df['Date'] = df['Time'].apply(getDateOnly)
df['Hour'] = df['Time'].apply(getHourOnly)
df['Minute'] = df['Time'].apply(getMinuteOnly)
df['Second'] = df['Time'].apply(getSecondOnly)
df
```

```python
def turnToNone(val):
    return None if val == '?' else val
for field in ['Age', 'Sentiment', 'Year', 'Month', 'Date', 'Hour', 'Minute', 'Date']:
    df[field] = df[field].apply(turnToNone)
df
```

```python
df = df.dropna()
df
```

```python
df = df[['Sentiment', 'Age', 'Year', 'Month', 'Date', 'Hour', 'Minute', 'Second']]
df
```
